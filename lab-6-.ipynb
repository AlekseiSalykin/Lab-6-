```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_excel("sample_data/films.xlsx", sheet_name='Sheet1')

openings = df["Год"].value_counts().head(5)
top_opening_names = openings.index
plt.bar(top_opening_names, openings)
plt.show()
```

```python
# Исходный набор данных
df1 = pd.read_excel("sample_data/grafik.xlsx", sheet_name="Sheet1")

# Настройка размеров подложки
plt.figure(figsize=(12, 7))

# Вывод графиков
plt.subplot(2, 2, 1)
plt.plot(df1['x'], df1['y1'], '-')
plt.subplot(2, 2, 2)
plt.plot(df1['x'], df1['y2'], '--')
plt.subplot(2, 2, 3)
plt.plot(df1['x'], df1['y3'], '-.')
plt.subplot(2, 2, 4)
plt.plot(df1['x'], df1['y4'], ':')
plt.show()
```

```python
import seaborn as sns
sns.pairplot(data=df1, vars=["x", "y1", "y2", "y3"])
plt.show()
```
